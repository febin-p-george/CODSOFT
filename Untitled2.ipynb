***TITANIC DATASET***

```python
import numpy as np
import pandas as pd
import statistics
```

**LOADING THE DATA**


```python
data=pd.read_csv('/content/Titanic-Dataset.csv')
data
```

```python
data.info
```

```python
data.isnull().sum()
```

**CLEANING THE DATA**

```python
data['Age'].fillna(data['Age'].mean(),inplace=True)
data['Cabin'].fillna(data['Cabin'].mode()[0],inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0],inplace=True)
data.replace({'Sex':{'male':0,'female':1},'Embarked':{'S':0,'C':1,'Q':2},},inplace=True)
data
```

```python
data['Cabin'].unique()
```

**SELECTING DIMENSIONS TO BE TRAINED WITH**

```python
X=data.drop(columns=['PassengerId','Name','Ticket','Survived','Cabin'],axis=1)
X
```

```python
Y=data['Survived']
Y
```

**CREATING AND TRAINING THE MODEL**

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
Xtrain.shape
```

```python
Ytrain.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X,Y)
```

```python
Ypred=model.predict(Xtest)
Ypred
```

**CHECKING THE ACCURACY**

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(Ytest,Ypred)
score
```

**TESTING WITH REAL WORLD DATA**

```python
prediction=model.predict([[3,0,22,1,0,7.25,0]])
prediction
```

**CONCLUSION**

The model is made with the Logistic Regression and the model has an accuracy of 78.2%. The model takes 80% of the provided data as the training data and 20% of the provided data as the test data.
